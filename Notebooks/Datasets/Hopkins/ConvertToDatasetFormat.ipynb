```python
import pandas as pd 

df = pd.read_excel('jh_dict_unicode.xls')
df.head()
```

```python
bad_lst = ['[', ']', '{', '}', '(', ')'] # these characters should not appear
ds_list = []
problem_lst = []

for index, row in df.iterrows():

    try:
        bo = row['Unnamed: 1']
        en = row['english1']

        lst = en.split(';')
        clean = [elt.strip()  for elt in lst if not any(bad in elt for bad in bad_lst)]

        for entry in clean:
            ds_list.append({'bo': bo, 'en': entry})
    
    except:
        try:
            bo = row['Unnamed: 1']
            en = row['english2']

            lst = en.split(';')
            clean = [elt.strip()  for elt in lst if not any(bad in elt for bad in bad_lst)]

            for entry in clean:
                ds_list.append({'bo': bo, 'en': entry})

        except:
            try:
                bo = row['Unnamed: 1']
                en = row['english-others']

                lst = en.split(';')
                clean = [elt.strip()  for elt in lst if not any(bad in elt for bad in bad_lst)]

                for entry in clean:
                    ds_list.append({'bo': bo, 'en': entry})
           
            except:
                
                problem_lst.append({'bo': bo, 'en': entry})
```
